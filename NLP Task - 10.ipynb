NLTK Pipeline

```python
import nltk
```

Text

```python
path_to_file="NLP-T10.txt"
with open(path_to_file) as f:
    contents = f.readlines()
```

```python
print(" ".join(contents))
texts=(" ".join(contents))
```

Power of Speech Tagging

```python
for text in texts:
    sentences=nltk.sent_tokenize(texts)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        tagged=nltk.pos_tag(words)
        print(tagged)
```
